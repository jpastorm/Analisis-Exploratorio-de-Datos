9.- Visualizar grafico de barras de la CANTIDAD DE LAS UNIVERSIDADES PRIVADAS Y \n PUBLICAS QUE POSTULARON POR EL LICENCIAMIENTO

```r
frec <- table(licenciamiento$TIPO_GESTION)
bp1  <- barplot(frec, main="CANTIDAD DE LAS UNIVERSIDADES PRIVADAS Y \n PUBLICAS QUE POSTULARON POR EL LICENCIAMIENTO", xlab="UNIVERSIDAD", ylab="Cantidad de Universidades",col = c("lightblue", 
        "lightcyan"),legend = rownames(frec),ylim=c(0,120))
```

10.-Visualizar gráficos de resumen de PORCENTAJE DE LAS UNIVERSIDADES PRIVADAS Y \n PUBLICAS QUE POSTULARON POR EL LICENCIAMIENTO: gráfico pie

```r
universidad <- table(licenciamiento$TIPO_GESTION)
piepercent<- round(100*universidad/sum(universidad), 1)
pie(universidad,piepercent,main ="PORCENTAJE DE LAS UNIVERSIDADES PRIVADAS Y \n PUBLICAS QUE POSTULARON POR EL LICENCIAMIENTO",col = rainbow(length(universidad)))
legend("topright",  c("PRIVADO", "PUBLICO"), cex = 0.8,
   fill = rainbow(length(universidad)))
```

11.-Visualizar la tabla y frecuencia con formato SAS del Estado de licenciamiento x Tipo de Gestion

12.-Visualizar la tabla y frecuencia con formato SPSS del Estado de licenciamiento x Tipo de Gestion

```r
Estado_Licenciamiento <- licenciamiento$ESTADO_LICENCIAMIENTO
Tipo_Gestion <- licenciamiento$TIPO_GESTION
CrossTable(Estado_Licenciamiento,Tipo_Gestion,format="SPSS")
```

13.-Visualizar el grafico en barra de Estado de licenciamiento x Tipo de Gestion 

```r
Estado_Licenciamiento <- licenciamiento$ESTADO_LICENCIAMIENTO
Tipo_Gestion <- licenciamiento$TIPO_GESTION
frec <- table(Estado_Licenciamiento,Tipo_Gestion)
frec
legendPlot=levels(Estado_Licenciamiento)
bp1  <- barplot(frec,beside = T, main="ESTADO DE LAS UNIVERSIDADES", xlab="TIPO DE UNIVERSIDADES", ylab="Cantidad de Universidades"
               ,col = rainbow(length(legendPlot)), ylim = c(0,70) )
legend("topright",  legend=legendPlot, cex = 0.8,fill = rainbow(length(legendPlot)))
```

14.-Visualizar la tabla y frecuencia con formato SPSS del Estado de licenciamiento x Tipo de Gestion x Periodo de Licenciamiento

```r
library(agricolae)
Estado_Licenciamiento <- licenciamiento$ESTADO_LICENCIAMIENTO
Periodo_Licenciamiento <- licenciamiento$PERIODO_LICENCIAMIENTO
Tipo_Gestion <- licenciamiento$TIPO_GESTION
frec <- table(Estado_Licenciamiento,Tipo_Gestion,Periodo_Licenciamiento)
frec
```

15.-Visualizar Histograma del periodo de licenciamiento

```r
(table.freq(
    hist(licenciamiento$PERIODO_LICENCIAMIENTO,
         breaks = "Sturges",
         col = "darkslategray2",
         main = "Histograma de los periodos de licenciamiento",
         xlab = "Periodos de licenciamiento",
         ylab = "Frecuencia"
        )))
```

16.-Visualizar Histograma y linea de distribucion del periodo de licenciamiento 

```r
histograma<- hist(licenciamiento$PERIODO_LICENCIAMIENTO,
         breaks = "Sturges",
         col = "darkslategray2",
         main = "Histograma de los periodos de licenciamiento",
         xlab = "Periodos de licenciamiento",
         ylab = "Frecuencia"
        )
polygon.freq(histograma,frequency=1,col="red")
```

17.-Visualizar linea de distribucion del periodo de licenciamiento 

```r
histograma1<- hist(licenciamiento$PERIODO_LICENCIAMIENTO,border = FALSE,
         main = "Histograma de los periodos de licenciamiento",
         xlab = "Periodos de licenciamiento",
         ylab = "Frecuencia"
        )
polygon.freq(histograma1,frequency=1,col="red")
```

17.-Visualizar grafico de barras de las Universidades por departamentos que han pedido el licenciamiento

```r
frec <- table(licenciamiento$DEPARTAMENTO_LOCAL)
frec
legendPlot <- licenciamiento$DEPARTAMENTO_LOCAL
bp1  <- barplot(
    frec,
    main="UNIVERSIDADES POR DEPARTAMENTOS \n QUE HAN PEDIDO LICENCIAMIENTO",
    xlab="",
    ylab="Cantidad",
    col = rainbow(length(legendPlot)),
    ylim=c(0,80),
    las=2,
    cex.names=0.6
)
```
